# Using Docling Convert PDF -> Markdown

```python
from docling.document_converter import DocumentConverter
```

```python
from docling.document_converter import DocumentConverter

pdf = "pdf1_gioi_thieu.pdf"

converter = DocumentConverter()
result = converter.convert(pdf)

doc = result.document
print(doc.export_to_markdown())
```

```python
output_md = "pdf1_gioi_thieu.md"
markdown_text = doc.export_to_markdown()

with open(output_md, "w", encoding="utf-8") as f:
    f.write(markdown_text)

print(f"Saved to {output_md}")
```

```python
print("Success")
```

# PyMUPDF4LLM

```python
import pymupdf4llm
md_text = pymupdf4llm.to_markdown("pdf1_gioi_thieu.pdf")
```

```python
import pymupdf4llm
llama_reader = pymupdf4llm.LlamaMarkdownReader()
llama_docs = llama_reader.load_data("pdf1_gioi_thieu.pdf")
```

```python
print(llama_docs)
```
