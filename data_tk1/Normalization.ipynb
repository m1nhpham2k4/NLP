```python
dict_viet_tat = {
    'sv': 'sinh viên',
    'gv': 'giảng viên',
    'đh': 'đại học',
    'cđ': 'cao đẳng',
    'tc': 'tín chỉ',
    'hk': 'học kỳ',
    'iuh': 'đại học công nghiệp thành phố hồ chí minh',
    'ktx': 'ký túc xá',
    'tp.hcm': 'thành phố hồ chí minh',
    'tphcm': 'thành phố hồ chí minh'
}

def replace_viet_tat(text):
    words = text.split()
    words = [dict_viet_tat.get(word, word) for word in words]
    return ' '.join(words)

def normalize_vietnamese_text(text):
    if not isinstance(text, str):
        return ""

    # chuẩn hóa Unicode về chuẩn NFC
    text = unicodedata.normalize('NFC', text)

    # chuyển toàn bộ thành chữ thường
    text = text.lower()

    # xử lý từ viết tắt
    text = replace_viet_tat(text)

    #xóa khoảng trắng thừa (tab, xuống dòng, khoảng trắng kép)
    text = re.sub(r'\s+', ' ', text).strip()

    #sdt
    text = re.sub(r'\b0\d{2,3}[\s.-]?\d{3,4}[\s.-]?\d{3,4}\b', '<PHONE>', text)

    # email
    text = re.sub(r'\S+@\S+', '<EMAIL>', text)

    # url
    text = re.sub(r'\b(?:https?://|www\.)\S+\b', '<URL>', text)
    text = re.sub(r'\b[A-Za-z0-9.-]+\.(vn|com|org|edu|gov|net)\b', '<URL>', text)

    return text
```

```python
import pandas as pd
df = pd.read_csv("truong_aug.csv")
score_cols = ["Score_1","Score_2","Score_3","Score_4","Score_5"]
aug_cols = ["Augmented_1","Augmented_2","Augmented_3","Augmented_4","Augmented_5"]

# tìm vị trí score lớn nhất
df["best_idx"] = df[score_cols].values.argmax(axis=1)

# lấy score lớn nhất
df["Best_Score"] = df[score_cols].max(axis=1)

# lấy text tương ứng
df["Best_Augmented"] = [
    df.iloc[i][aug_cols[idx]]
    for i, idx in enumerate(df["best_idx"])
]

# lọc score >= 0.65
df_filtered = df[df["Best_Score"] >= 0.65].copy()

# giữ lại cột cần thiết
df_filtered = df_filtered[[
    "Contents",
    "Labels",
    "Best_Augmented",
    "Best_Score"
]]

print(df_filtered.head())
```
