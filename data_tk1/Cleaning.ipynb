#Hàm làm sạch văn bản quy chế quy định

```python
import re

def clean_text(text):

    if not isinstance(text, str):
        return text

    text = text.strip()

    # lowercase
    text = text.lower()

    # replace url
    text = re.sub(r'http[s]?://\S+', '[URL]', text)

    # remove list markers
    text = re.sub(r'^\s*(\d+\.|[IVXLCDM]+\.|[a-zA-Z]\.|[a-zA-Z]\)|[-+–])\s*', '', text)

    # normalize dash
    text = text.replace('–', '-')

    # remove special characters
    text = re.sub(r'[\"“”…<>*&=]', '', text)

    # normalize spaces
    text = re.sub(r'\s+', ' ', text)

    return text.strip()
```

```python
df['Cleaned_Text'] = df['Thong_Tin'].apply(clean_text) #làm sạch dữ liệu nhãn Quy_Che_Quy_Dinh
df
```
