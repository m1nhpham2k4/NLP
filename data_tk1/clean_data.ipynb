```python
import pandas as pd
from underthesea import word_tokenize
import re

# =========================
# 1. Load dataset
# =========================

df = pd.read_csv("data.csv")

print("Original dataset size:", df.shape)


# =========================
# 2. Remove missing values
# =========================

df = df.dropna(subset=["Thong_Tin"])


# =========================
# 3. Remove duplicates
# =========================

df = df.drop_duplicates(subset=["Thong_Tin"])


# =========================
# 4. Text cleaning function
# =========================

def clean_text(text):

    text = str(text)

    # lowercase
    text = text.lower()

    # replace url
    text = re.sub(r"http\S+|www\S+", "<url>", text)

    # replace phone number
    text = re.sub(r"\d{3,}", "<num>", text)

    # remove punctuation
    text = re.sub(r"[^\w\s<>]", " ", text)

    # normalize spaces
    text = re.sub(r"\s+", " ", text)

    return text.strip()


df["clean_text"] = df["Thong_Tin"].apply(clean_text)


# =========================
# 5. Truncate long text
# =========================

def truncate_text(text, max_words=64):

    words = text.split()

    if len(words) > max_words:
        return " ".join(words[:max_words])

    return text


df["clean_text"] = df["clean_text"].apply(lambda x: truncate_text(x, 64))


# =========================
# 6. Recalculate length
# =========================

df["length"] = df["clean_text"].apply(lambda x: len(x.split()))


# =========================
# 7. Remove duplicates again
# =========================

df = df.drop_duplicates(subset=["clean_text"])


# =========================
# 8. Reset index
# =========================

df = df.reset_index(drop=True)


# =========================
# 9. Save cleaned dataset
# =========================

df["tokenized"] = df["clean_text"].apply(lambda x: word_tokenize(x, format="text"))


df.to_csv("dataset_cleaned.csv", index=False, encoding="utf-8-sig")

print("Cleaned dataset size:", df.shape)
print("Saved file: dataset_cleaned.csv")
```
